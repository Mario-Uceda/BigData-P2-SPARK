```python
#Solo productos vendidos en españa
df_españa = df_pequeño.where(df_pequeño.countries_en == 'Spain')
#Elimino la columna de pais porque ya se que todos los productos son de españa
df_españa=df_españa.drop('countries_en')
#Eliminamos los productos que no indiquen donde se venden
df_tiendas_españa = df_españa.dropna(subset=['stores'])
nproductos = df_tiendas_españa.count()
print('Número de productos que se venden en España: '+str(nproductos))
```

```python
#Eliminamos los productos que no indiquen ninguna de las siguinetes características
df_caracteristicas = df_tiendas_españa.dropna(thresh=1,subset=('vitamin-c_100g','vitamin-b9_100g','vitamin-b12_100g','iron_100g'))
nproductos = df_caracteristicas.count()
print('Número de productos que indican alguna de las siguientes características: '+str(nproductos))
```
